```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
anime = pd.read_csv("anime.csv")
```

```python
anime = anime[['anime_id', 'name', 'genre']].dropna()
anime.drop_duplicates(inplace=True)
```

```python
anime['genre'] = anime['genre'].fillna('').str.replace(' ', '')
```

```python
cv = CountVectorizer(max_features=3000, stop_words='english')
vectors = cv.fit_transform(anime['genre'])

similarity = cosine_similarity(vectors)
```

```python
def recommend(anime_name):
    idx = anime[anime['name'] == anime_name].index[0]
    distances = similarity[idx]

    anime_list = sorted(list(enumerate(distances)),
                         reverse=True,
                         key=lambda x: x[1])[1:6]

    for i in anime_list:
        print(anime.iloc[i[0]]['name'])
```

```python
recommend("Naruto")
```
